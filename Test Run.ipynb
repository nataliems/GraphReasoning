```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
from transformers import AutoTokenizer, AutoModelForCausalLM
import os
from GraphReasoning import *

# Specify the local model path
model_path = os.path.abspath("models/neuralmagic-Meta-Llama-3-70B-Instruct-quantized.w8a16/snapshots/860ed63dce2e0711e0d6bacedf00b02c3ccd4e6e")

# Load the model and tokenizer with authentication
tokenizer = AutoTokenizer.from_pretrained(model_path, local_files_only=True)
model = AutoModelForCausalLM.from_pretrained(model_path, local_files_only=True)
```

```python
text = """
Silk is a strong natural fiber used to catch prey in a web. Beta-sheets control its strength. 
The tensile properties of silk make it valuable in biomedical applications.
"""
```

```python
graph_HTML, graph_GraphML, G, net, output_pdf = make_graph_from_text(
    txt=text,
    generate=model.generate(),                # Pass the Mistral-based generate function
    include_contextual_proximity=True, # Include co-occurrence relationships
    graph_root="example_graph",        # Prefix for saved graph files
    chunk_size=250,                    # Maximum chunk size for text splitting
    chunk_overlap=50,                  # Overlap between chunks
    repeat_refine=1,                   # Number of refinement iterations
    verbatim=True,                     # Print debug outputs
    save_HTML=True,                    # Save HTML visualization
    save_PDF=False                     # Save PDF visualization
)
```
